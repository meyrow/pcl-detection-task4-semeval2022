```python
# Load the model from the file
import joblib
svm_from_joblib = joblib.load('classifier_svm.pkl')
```

```python
import pandas as pd
dataExam = pd.read_pickle("allTest.pkl")
dataExam.head(10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 2000)
X_input = cv.fit_transform(dataExam['sentence'].values).toarray()
```

```python
# use model to predict
svm_pred = svm_from_joblib.predict(X_input)
```

```python
svm_pred
```

```python
svm_pred = svm_pred.astype(int)
svm_pred
```

```python
for pi in svm_pred:
  print(pi)
```

```python
# helper function to save predictions to an output file
def labels2file(p, outf_path):
	with open(outf_path,'w') as outf:
		for pi in p:
			outf.write(str(pi)+'\n')
```

```python
import os
# first, we need to create the res/ and ref/ folders, which the evaluator expects
!mkdir res
```

```python
labels2file(svm_pred, os.path.join('res/', 'svm_task1.txt'))
```
